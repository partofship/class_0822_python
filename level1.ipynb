__[문제5]__

str = "야호야호야호호야호야야오햐오햐오햐오"

str을 3가지 이상의 다른 문자열 슬라이싱을 이용하여 확인해보세요!

출력 예)

처음 5글자: 야호야호야


```python
example_str = "야호야호야호호야호야야오햐오햐오햐오"

print(f"처음 5글자: {example_str[:5]}")
print(f"5글자~9글자: {example_str[4:9]}")
print(f"10글자 이후: {example_str[9:]}")
```

__[문제7]__

변수명 전화번호에 들어 있는 번호를 모두 붙여서 출력되도록 코드를 작성해주세요.

<br>

phone = "010-1234-5678"

출력 예시 : "01012345678"

<br>

__Tips__
- split() 함수는 문자열을 특정 구분자를 기준으로 나누어 리스트로 반환합니다.


```python
phone_num = "010-1234-5678"
result= "".join(phone_num.split("-"))
print(result)
```
